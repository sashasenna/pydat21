## 2. Факультативы (множества)
Группа из 3 студентов пишет заявки на желаемые факультативы из списка: английский, немецкий, право, математика, сольфеджио. Факультатив откроют, если на него запишутся все студенты. Каждый студент может выбрать минимум один и максимум три факультатива. Нужно посчитать количество факультативов, которые откроются.

*Пример*

**Ввод:**

английский сольфеджио право  
математика сольфеджио  
немецкий право

**Вывод:**

0

**Ввод:**

математика немецкий право  
математика немецкий  
немецкий право математика 

**Вывод:**  
2

```python
S1 = set(input().split())
S2 = set(input().split())
S3 = set(input().split())
S4 = S1 & S2 & S3
print(len(S4))
```

```python
print(len(set(input().split()) & set(input().split()) & set(input().split())))
```

*Пример*  
**Ввод:**

математика немецкий право

право

**Вывод:**

```python
s = set(input().split()) | set(input().split())
s.remove(input())
print(", ".join(s))
```

### 4. Сколько чисел совпадает? (условия)

Даны три целых числа. Определите, сколько среди них совпадающих. Программа должна вывести одно из чисел: 3 (если все совпадают), 2 (если два совпадает) или 0 (если все числа различны).

**Формат ввода**  
Вводятся три целых числа

**Формат вывода**  
Целое число

```python
number1 = input()
number2 = input()
number3 = input()
S = {number1, number2, number3}

if len(S) == 1:
    print("Все числа совпадают")
elif len(S) == 2:
    print("Совпадает два числа")
else:
    print("Все числа разные")
```

## 5. Распродажа

В магазине проходит акция:
* На все товары дешевле 1000 рублей скидка 15%
* На все товары дороже 1000, но дешевле 5000 рублей скидка 20%
* На все товары дороже 5000 рублей скидка 25%

**Ввод**  
Целое неотрицательное число - цена товара в рублях

**Вывод**  
Целое неотрицательное число - скидка на товар в рублях

```python
price = int(input())

if price < 1000:
    print(f"Ваша скидка 15%. Сумма к оплате: {price * 0.85 : 0.2f}")
elif price < 5000:
    print(f"Ваша скидка 20%. Сумма к оплате: {price * 0.8 : 0.2f}")
else:
    print(f"Ваша скидка 25%. Сумма к оплате: {price * 0.75 : 0.2f}")
```

## 6. Какой-то яблочный бизнес...

После очередного успеха на олимпиаде Никита и Кирилл решили открыть собственный бизнес. Бизнес не простой, а золотой — продажа яблок сорта Golden.

Миллионер Семён решил порадовать любимую мамочку — купить ей вкусных яблок. Он пришёл на импровизированный рынок НиК, где работали Никита и Кирилл. Семён решил купить все яблоки, благо денег у него хватает.

Продав все яблоки, Никита и Кирилл стали считать прибыль. Известно, что Никита продал A яблок по M тугриков, а Кирилл продал B яблок по N тугриков. К сожалению, с математикой у ребят плохо, поэтому они хотят узнать у вас, кто же получил бОльшую прибыль.

Числа A,B,M,N вводятся пользователем.

Как должно работать:

<table>
	<tr><th>Что выводит программа?</th><th>Что вводим?</th></tr>
	<tr><td>A:<br>M:<br>B:<br>N:<br>Никита заработал больше</td><td>5<br>1000<br>6<br>500</td></tr>
	<tr><td>A:<br>M:<br>B:<br>N:<br>Кирилл заработал больше</td><td>5<br>1000<br>10<br>8000</td></tr>
	<tr><td>A:<br>M:<br>B:<br>N:<br>Никита и Кирилл заработали поровну</td><td>5<br>1000<br>10<br>500</td></tr>
</table>

```python
A = int(input("Сколько яблок продал Никита: "))
M = float(input("По какой цене продал яблоки Никита: "))
B = int(input("Сколько яблок продал Кирилл: "))
N = float(input("По какой цене продал яблоки Кирилл: "))

if A * M > B * N:
    print("Никита заработал больше")
elif A * M < B * N:
    print("Кирилл заработал больше")
else:
    print("Никита и Кирилл заработали поровну")
```

## 7. Какой-то яблочный бизнес...

Красная Шапочка часто навещает свою бабушку. Но она очень боится, что рано или поздно ее бабушку опять навестит волк. Поэтому она решила договориться с Лесничим об охране бабушки. Лесничий согласился охранять бабушку за 10 пирожков.

Узнав об этом, волк сказал Красной Шапочке, что ей совершенно незачем тратить пирожки на Лесничего. За половину тех пирожков, которые Красная Шапочка несет бабушке, Волк пообещал не трогать ее.

Мама испекла несколько пирожков, и попросила Красную Шапочку отнести их бабушке. Требуется определить, кому выгоднее платить пирожки за спокойствие бабушки.

<table>
	<tr><th>Что выводит программа?</th><th>Что вводим?</th></tr>
	<tr><td>Кому платить:<br>Лесничий</td><td>50</td></tr>
	<tr><td>Кому платить:<br>Волк</td><td>5</td></tr>
	<tr><td>Кому платить:<br>Волк</td><td>19</td></tr>
	<tr><td>Кому платить:<br>Лесничий</td><td>20</td></tr>
</table>

```python
# Решение с дополнительными переменными

pie = int(input())
wolf = pie - (pie / 2)
forest_ranger = pie - (pie - 10)

if wolf < forest_ranger:
    print("Волк")
else:
    print("Лесничий")
```

```python
# Решение без дополнительных переменных

pie = int(input())
wolf = pie - (pie / 2)
forest_ranger = pie - (pie - 10)

if pie - (pie / 2) < pie - (pie - 10):
    print("Волк")
else:
    print("Лесничий")
```

## 8. Невероятное везение...

Напишите программу, в которой пользователю предлагается с первого раза угадать число от 1 до 10.

Если пользователь угадал, программа поздравляет его с выигрышем!

```python
import random

participants_number = int(input("Введите целое число от 1 до 10: "))
random_number = random.randint(1, 10)

if participants_number == random_number:
    print("Ура! Поздравляем, Вы выиграли!")
else:
    print("Вам не повезло... Попробуйте еще раз!")
```
